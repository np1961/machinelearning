```python
from time import sleep
from os import system
import numpy as np
from pandas import DataFrame
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot
```

```python
class linear_regression:
    
    def __init__(self,
                 in_values='in_values',
                 out_values='out_values',
                 figsize=(5,5),
                 learning_rate=0.1):
        
        self.in_values=in_values
        self.out_values=out_values
        self.figsize=figsize
        
        self.eps=0.19
        self.learning_rate=learning_rate
        
        
        try:
            self.x_this=np.random.randint(
                        min(self.in_values),
                        max(self.in_values)
                                     )

            self.y_this=self.function(
                                 x=self.x_this,
                                 config='parabol')
            print('linear model init')
        except:
            print('template object init')
        
        
    
    def function(self,x, config='parabol'):
        if config=='parabol':
            return x**2
        else:
            print('error')
    
    def grad_function(self,x, config='parabol'):
        
        if config=='parabol':
            return 2*x
        else:
            print('error')
        
    def visual_2d_step(self, update=False):
        pyplot.figure(figsize=self.figsize)
        pyplot.plot(self.in_values,
                    self.out_values,
                    color='g')
        
        pyplot.scatter(self.x_this, self.y_this, c='b', linewidths=10)
        
        print(self.x_this, self.y_this)
        
        
        x_new=self.x_this-self.grad_function(
                        x=self.x_this)*self.learning_rate
        
        y_new=self.function(x=x_new)
        
        print(x_new,y_new)
        
        pyplot.scatter(x_new, y_new, c='r', linewidths=6)
        pyplot.plot([self.x_this, x_new],[self.y_this, y_new] , 'y')
        pyplot.show()
        sleep(1) 
        if update:
            if abs(self.x_this-x_new)<self.eps:
                return False
            else:
                self.x_this=x_new
                self.y_this=y_new
        else:
            return True
        
        
```

```python
x=np.linspace(-5,5,30)
```

```python
y=linear_regression().function(x=x)
```

```python
model=linear_regression(in_values=x,
                        out_values=y)
```

```python
while True:
    model.visual_2d_step(True)
```
