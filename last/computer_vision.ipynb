```python
import numpy as np
from pandas import read_csv
from pandas import DataFrame
from matplotlib import pyplot
from sklearn.preprocessing import StandardScaler, normalize
import easyocr
import cv2 as computer_vision
from warnings import filterwarnings as message
from random import random as random
message('ignore')
```

```python
image_path='/home/np_1961/PycharmProjects/pythonProject2/14.png'
image=computer_vision.imread(image_path)
reader=easyocr.Reader(['en'],gpu=False)
results=reader.readtext(image_path)
text=''
for matrix, result, score in results:
    print(result, '_____', str(int(score*100))+'%' )

top_left=tuple(results[0][0][0])
botton_right=tuple(results[0][0][2])
text=results[0][1]
font=computer_vision.FONT_HERSHEY_SIMPLEX
image=computer_vision.imread(image_path)
image=computer_vision.rectangle(image,top_left,
                                botton_right,(0,255,0),5)
image=computer_vision.putText(image,text,top_left,font,
                             .5, (255,0,255), 2, 
                                 computer_vision.LINE_AA)
pyplot.imshow(image)
pyplot.show()
```

```python
image_path='/home/np_1961/one.jpeg'
image=computer_vision.imread(image_path)
reader=easyocr.Reader(['en'],gpu=False)
results=reader.readtext(image_path)
text=''
for result in results:
    top_left=tuple([int(element) for element in result[0][0]])
    botton_right=tuple([int(element) for element in result[0][2]])
    text += result[1]+ ' '
    font=computer_vision.FONT_HERSHEY_SIMPLEX
    image=computer_vision.rectangle(image,top_left,
                                   botton_right, 
                                   (0,255,0), 5)
    image=computer_vision.putText(image,text,
                                 top_left, font,
                                 0.2, (255,0,0), 1,
                                 computer_vision.LINE_AA)
pyplot.figure(figsize=(10,10))
pyplot.imshow(image)
print(text)
pyplot.show()
```
